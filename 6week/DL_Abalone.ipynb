нҡҢкё° лӘЁлҚёмқҖ м „ліөмқҳ кі лҰ¬ мҲҳк°Җ лӘҮ к°ңмқёк°Җ?
['Rings']лҘј мқҙмҡ©н•ҙ мӢ мІҙм§ҖмҲҳ (X)лҘј мһ…л Ҙл°ӣм•„ лӮҳмқҙлҘј мһҳ л§һм¶”лҠ” кІғмқҙ лӘ©н‘ң

л¶„лҘҳ лӘЁлҚёмқҖ м „ліөмқҳ мғқм•  мЈјкё°к°Җ м–ҙл””м—җ мҶҚн•ҳлҠ”к°Җ?


```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from tensorflow.keras import layers, models

#мғҒкҙҖкҙҖкі„ нҷ•мқёмқ„ мң„н•ң sebornкіј matplot
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. лҚ°мқҙн„° лЎңл“ң (abalone)
path_aba = "/content/drive/MyDrive/б„Җб…Ўб„Һб…ҘбҶ«б„ғб…ўб„’б…ЎбҶЁб„Җб…ӯ/2026-1/б„Ӣб…өбҶ«б„Җб…©бҶјб„Ңб…өб„Ӯб…ібҶјб„Җб…ўб„…б…©бҶ«/Colab Notebooks/2. б„ғб…Ұб„Ӣб…өб„җб…Ҙ/abalone.csv"
df = pd.read_csv(path_aba, index_col=0)
df
```

```python
# кІ°мёЎм№ҳ нҷ•мқё
df.isnull().sum()
```

кІ°мёЎм№ҳк°Җ м—ҶмңјлҜҖлЎң ['Rings']мҷҖмқҳ мғҒкҙҖкҙҖкі„ нҷ•мқён•ҳкё°

```python
# мғҒкҙҖкҙҖкі„ л¶„м„қмқ„ мң„н•ҙ мӣҗ н•« мқёмҪ”л”© мҡ°м„  м§„н–ү
df_encoded = pd.get_dummies(df, columns=['Sex'])
corr_mat=df_encoded.corr()
# нҺён•ҳкІҢ ліј мҲҳ мһҲлҸ„лЎқ ascending=False
corr_mat['Rings'].sort_values(ascending=False)
```

```python
# л¶„м„қкІ°кіјлҘј л°”нғ•мңјлЎң X(нҠ№м§•) y(м •лӢө) лӮҳлҲ„кё°
y_reg = df_encoded['Rings'].values

# Rings л°ҳл“ңмӢң м ңмҷё, idлҠ” лҜёлҰ¬ м ңкұ°н•Ё
X = df_encoded.drop(['Rings'], axis=1)
```

```python
# мҠӨмјҖмқјл§Ғ (л„ҳнҢҢмқҙлЎң мһҗлҸҷ ліҖнҷҳ)
scaler_reg = StandardScaler()
X_reg_scaled = scaler_reg.fit_transform(X)

# train / test л¶„н• 
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_reg_scaled, y_reg, test_size=0.2, random_state=0)
```

```python
# л”Ҙлҹ¬лӢқ нҡҢкё°
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train_r.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mse']
)

history = model.fit(
    X_train_r, y_train_r,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    verbose=1
)
```

Regression - ['Sex']лҘј мӢ мІҙм§ҖмҲҳлЎң кө¬л¶„н•ҳкё°

```python
# л¶„м„қкІ°кіјлҘј л°”нғ•мңјлЎң X(нҠ№м§•) y(м •лӢө) лӮҳлҲ„кё°
y = df['Sex']
# SexлҠ” л°ҳл“ңмӢң м ңмҷё, нҠ№м§•мқҖ м№ҳмҲҳлҘј к°Җм§„ лӘЁл“  м№јлҹјмқҙ нҠ№м§•мқҙ
X = df.drop('Sex', axis=1)
```

```python
# yмқҳ мҲңм„ңлҘј лқјлІЁл§Ғ, мӣҗ н•« мқёмҪ”л”©
classes = sorted(y.unique()) # ['F', 'I', 'M'] лһңлҚӨн•ң мҲңм„ңлЎң л“Өм–ҙмҷ”мқ„ л•Ң м°©мҳӨк°Җ м—ҶлҸ„лЎқ к°Ғмһҗмқҳ мң„м№ҳлҘј м •н•ҙл‘ !
y_encoded = pd.get_dummies(y).reindex(columns=classes, fill_value=0).values
```

```python
# мҠӨмјҖмқјл§Ғ (л„ҳнҢҢмқҙлЎң мһҗлҸҷ ліҖнҷҳ)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# train/test л¶„н• 
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
```

```python

# л”Ҙлҹ¬лӢқ л¶„лҘҳ
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dropout(0.3), # кіјм Ғн•© л°©м§Җ
    layers.Dense(32, activation='relu'),
    layers.Dense(y_train.shape[1], activation='softmax') # 3к°ңлЎң л¶„лҘҳ (F, I, M)
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    verbose=1
)
```

```python
# м„ұм Ғмқ„ нҷ•мқён•ҳмһҗ!
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"мөңмў… м„ұлі„ нҢҗлі„ м •нҷ•лҸ„: {acc * 100:.2f}%")
```
