데이터의 각 테이블 분석



```python
# 데이터 관리 도구
import pandas as pd  #
import numpy as np   #

# 데이터 전처리 도구
from sklearn.model_selection import train_test_split # 데이터를 학습용/검증용으로 나눔
from sklearn.preprocessing import LabelEncoder      # 문자를 숫자로 바꿈
from sklearn.preprocessing import StandardScaler    # 숫자의 크기(Scale)를 일정하게 맞춤

# AI 모델(알고리즘) 도구
from sklearn.linear_model import LogisticRegression  # 선형 분류(LR)
from sklearn.tree import DecisionTreeClassifier       # 의사결정나무(DT)
from sklearn.ensemble import RandomForestClassifier  # 랜덤 포레스트(RF)
from sklearn.neighbors import KNeighborsClassifier   # K-최근접 이웃(KNN)

# 성능 점수판 도구
from sklearn.metrics import accuracy_score #정답의 정확도를 계산
```

```python
#df.drop([...])	불필요한 정보 제거
#fillna()	빈칸 채우기
#le.fit_transform()	라벨 인코딩
#StandardScaler()	데이터 표준화
```

```python
# 1. 데이터 로드
path = "/content/drive/MyDrive/가천대학교/2026-1/인공지능개론/Colab Notebooks/2. 데이터/titanic.csv"
df = pd.read_csv(path)
df
```

```python
# 2. 필요 없는 칼럼 제거
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, errors='ignore')
df
# 'ignore' 칼럼이 없을 경우 pass 한다! 중복 실행으로 인한 오류 해결
```

```python
# 3. 결측치(NaN, Null) 처리
df['Age'] = df['Age'].fillna(df['Age'].median()) # 나이는 중앙값 채움
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0]) # 항구는 최빈값을 채움
df
```

```python
# 4. Label Encoding (글자 -> 숫자 변경)
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex']) # male:1, female:0 (반대로도 ok)
df['Embarked'] = le.fit_transform(df['Embarked'])
df
#
```

```python
# 5. x, y로 (특징과 정답) 분리
X = df.drop('Survived', axis=1)
y = df['Survived']
X, y
```

```python
# 6. 데이터 스케일링(KNN에 맞추기)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 6. 학습(8:2)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=0)
```

```python
# 모델 생성
models = [LogisticRegression(), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier()]

# 학습 및 결과 출력(for문으로 통합)
for model in models:
    model.fit(X_train, y_train)
    name = model.__class__.__name__ # class의 name 담기
    print(f"{name} 모델의 결과: {accuracy_score(y_test, model.predict(X_test))}")
```

```python
# 7. 모델의 정확도와 사망확률은 다르다 -> 확률로 도출
for model in models:
    # 1. 모델 학습
    model.fit(X_train, y_train)

    # 2. 모델 이름 가져오기
    name = model.__class__.__name__

    # 3. 정확도(성적표) 계산
    acc = accuracy_score(y_test, model.predict(X_test))

    # 4. 개별 승객의 '확률' 계산 (중요!)
    # predict_proba는 [사망확률, 생존확률] 두 값을 줍니다.
    probs = model.predict_proba(X_test)

    # 결과 출력
    print(f"[{name}]")
    print(f"- 전체 정확도: {acc:.4f}")
    # probs[0][0]은 첫 번째 승객의 사망 확률, probs[0][1]은 생존 확률입니다.
    print(f"- 첫 번째 승객 확률: [사망 {probs[0][0]:.2%}, 생존 {probs[0][1]:.2%}]")
    print("-" * 30)
```
