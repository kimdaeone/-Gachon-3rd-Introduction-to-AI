```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split


# 데이터베이스(내장)
from sklearn import datasets
data = datasets.load_wine()
data
```

```python
import pandas as pd
import numpy as np

y=data.target

X=data.data

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

ly_preds_lr = model_lr.predict(X_test)

print ("\n--- Linear Regression Regression ---")
print('평균제곱오차', mean_squared_error(ly_preds_lr, y_test))


model_dt = DecisionTreeRegressor()
model_dt.fit(X_train, y_train)

ly_preds_dt = model_dt.predict(X_test)

print ("\n--- DecisionTree Regression ---")
print('평균제곱오차', mean_squared_error(ly_preds_dt, y_test))


model_rf = RandomForestRegressor()
model_rf.fit(X_train, y_train)

ly_preds_rf = model_rf.predict(X_test)

print ("\n--- RandomForest Regression ---")
print('평균제곱오차', mean_squared_error(ly_preds_rf, y_test))


model_kn = KNeighborsRegressor()
model_kn.fit(X_train, y_train)

ly_preds_kn = model_kn.predict(X_test)

print ("\n--- KNeighborsRegressor Regression ---")
print('평균제곱오차', mean_squared_error(ly_preds_kn, y_test))
```
